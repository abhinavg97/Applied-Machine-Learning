## Importing Libraries

```python
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

## Loading the dataset

```python
iris = datasets.load_iris()
```

## Finding the PCA using the in-built function

```python
X_tsne = TSNE(learning_rate=200,random_state=1).fit_transform(iris.data)
```

## Plotting the result 

```python
plt.figure(figsize=(10, 5))
plt.title('TSNE for IRIS dataset')
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=iris.target, label=iris.target)

plt.show()
```
