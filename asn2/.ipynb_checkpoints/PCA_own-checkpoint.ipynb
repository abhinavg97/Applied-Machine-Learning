## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Defining the PCA finding function 

```python
# Defining the function for finding the PCA for a given dataset
def PCA(data, dim_num=2):
    m, n = data.shape
    mean = data.mean(axis=0)

    # Center the data around the mean
    data -= mean

    # Calculating the covariance matrix
    cov_matrix = np.cov(data, rowvar=False)

    # Calcuating the eigen values and eigen vectors for the data matrix
    eig_vals, eig_vecs = np.linalg.eigh(cov_matrix)
#     print eig_vals.shape
#     print eig_vals.shape
    # Sorting the eigen values in descending order in order to see which of the values describes the data set the best
    x = np.argsort(eig_vals)[::-1]
    # Sorting the eigen vectors with respect to eigen values
    eig_vecs = eig_vecs[:,x]

    eig_vals = eig_vals[x]

    # From the sorted list, we are selecting the first n number of dimensions(here dim_num = 2) for both eigen values and eigen vectors
    eig_vecs = eig_vecs[:,:dim_num]
    
#     print eig_vecs.shape, data.shape
    # Returning the eigen values, eigen vector
    return np.dot(eig_vecs.T, data.T).T, eig_vals, eig_vecs
```

## Loading the iris dataset

```python
# Loading the dataset
dataset = np.genfromtxt('Iris.csv', delimiter=',')
```

## Getting the PCA values for the iris dataset 

```python
X = dataset[:,0:4]

Matr, eig_vals, eig_vecs = PCA(X)

y = np.array([0]*150)
y[50:100] = 1
y[100:150] = 2
```

## Plotting the result

```python
plt.figure(figsize=(10,5))
colors = ['navy', 'turquoise', 'darkorange']
target_names = ["Iris-virginica", "Iris-versicolor", "Iris-setosa"]
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(Matr[y == i, 0], Matr[y == i, 1], color=color, alpha=.8, lw=lw, label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')

plt.show()
```

## T SNE from in-built library

```python
from sklearn.manifold import TSNE
from sklearn import datasets

iris = datasets.load_iris()

X_tsne = TSNE(learning_rate=200,random_state=1).fit_transform(iris.data)

plt.figure(figsize=(10, 5))
plt.title('TSNE for IRIS dataset')
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=iris.target, label=iris.target)

plt.show()
```

## SwissRoll dataset

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

from sklearn import datasets
from sklearn.manifold import TSNE

X, color = datasets.make_swiss_roll(n_samples=1500, noise=0.0, random_state=None)

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

ax.scatter(X[:,0], X[:,1], X[:,2], c=color, cmap=plt.cm.spectral)

ax.set_title("Original data")

plt.show()


X_tsne = TSNE(learning_rate=100, random_state=1).fit_transform(X)
X_pca, eig_vals, eig_vecs = PCA(X)


plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=color)
plt.title('TSNE for SwissRoll dataset')
plt.subplot(122)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=color)
plt.title('PCA for SwissRoll dataset')


plt.show()
```
